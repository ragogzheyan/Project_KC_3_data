```python
import pandas as pd 
import numpy as np 
```

```python
# Method 2 same project
data_1 = {
         'key': ['A0', 'A1', 'A2', 'A3', 'A4','A5','A7','A8','A9','A10','A11','A12'],
         'cost': ['room_board', 'state_tuition', 'clothes', 'entertainment', 'unforeseen_cost', 'miscellaneous',
                  'savings', 'parents', 'job', 'financial_aid', 'other_cost','miscellaneous'], 
         'freshman': [3290.00, 7967.00, 490.00, 635.00, 325.00, 0, 1300.00, 5080.97, 1450.00, 4200.00, 300.00, np.nan]
}
        
data_1 = pd.DataFrame(data_1)

data_1.head(12)

data_1
```

```python
data_2 = {
         'key': ['A0', 'A1', 'A2', 'A3', 'A4','A5','A7','A8','A9','A10','A11','A12'], 
         'sophomore': [3454.50, 7967.00, 514.50, 666.75, 341.25, np.nan,1300.00, 5197.97, 1522.50, 4410.00, 380.00, 0],
         'junior': [3627.23, 7967.00, 540.22, 700.09, 358.31, 0, 1314.00, 5320.82, 1598.64, 4630.50, 450.00, np.nan],
         'senior': [3808.59, 7967.00, 567.24, 735.09, 376.23, np.nan,1352.20, 5449.81, 1678.56, 4862.03, 500.00, 0]
         }

data_2 = pd.DataFrame(data_2)

data_2.head(12)

data_2 
```

```python
merged_df = data_1.merge(data_2, how='inner', on='key').fillna(0)

merged_df
```

```python

merged_df.loc[:,'total'] = merged_df.sum(numeric_only=True, axis=1)

df1 = merged_df.fillna(0)

df1
```

```python

df1.loc[len(df1.index)] = df1.loc[0:5,['freshman','sophomore','junior','senior','total']].sum(numeric_only=True,axis = 0)
df1.at[12,'cost'] = 'total_expenses'

df1.loc[len(df1.index)+1] = df1.loc[6:11,['freshman','sophomore','junior','senior','total']].sum(numeric_only=True,axis = 0)
df1.at[len(df1.index),'cost'] = 'total_resources'


df1.fillna(0)
```
